## DATA PREPROCESSING AND DATA CLEANING

```python
import pandas as pd
import numpy as np


import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('MALWARE.csv')
del df['hash']
del df['normal_prio']
del df['policy']
del df['vm_pgoff']
del df['usage_counter']
del df['cgtime']
del df['signal_nvcsw']
del df['gtime']
del df['lock']
del df['hiwater_rss']
del df['task_size']
del df['cached_hole_size']
del df['nr_ptes']

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

var = ['classification']
  
for i in var:
    df[i] = le.fit_transform(df[i]).astype(int)
```
