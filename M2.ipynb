## DATA VISUALIZATION AND DATA ANALYSIS

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('TCS1.csv',usecols=["Open","High","Low","Volume","Close"])
df.head()
```

```python
df.columns
```

```python
df = df.dropna()
```

```python
df=df.drop_duplicates()
```

```python
df.tail()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(df['Open'],color='purple')

plt.subplot(1,2,2)
plt.hist(df['Close'],color='tomato')
```

```python
df.hist(figsize=(15,55),layout=(15,4), color='lightblue')
plt.show()
```

```python
df['Volume'].hist(figsize=(10,5),color='lightgreen')
```

```python
plt.scatter(df['High'],df['Volume'], color='pink') # scatter, plot, triplot, stackplot
```

```python
plt.boxplot(df['Open']) 
```

```python
sns.boxenplot(df['High'], color='purple') 
# barplot, boxenplot, boxplot, countplot, displot, distplot, ecdfplot, histplot, kdeplot, pointplot, violinplot, stripplot
```

```python
fig,ax=plt.subplots(figsize=(30,20))
sns.heatmap(df.corr(),annot=True,fmt='0.2%',cmap='autumn',ax=ax)
```
